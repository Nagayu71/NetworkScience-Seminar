```python
import networkx as nx
import numpy as np
import quantecon as qe
import matplotlib.pyplot as plt
```

# 1.4.1 Unweighted Directed Graphs

## 1.4.1.2 Digraphs in Networkx
Figure 1.9: A digraph of classes
- `p`: poor
- `m`: middle class
- `r`: rich

```python
# Create an empty DiGraph object
G_p = nx.DiGraph()
```

```python
# Populate G_p with nodes and edges
edge_list = [
    ('p', 'p'),
    ('m', 'p'), ('m', 'm'), ('m', 'r'),
    ('r', 'p'), ('r', 'm'), ('r', 'r'),
]
```

```python
# Add edges from edge_list
# Instead of the below, you can use DiGraph.add_edges_from() method.
for e in edge_list:
    u, v = e
    G_p.add_edge(u, v)
```

```python
# Plots G_p
fig, ax = plt.subplots()
nx.draw_spring(G_p, ax=ax, node_size=500, with_labels=True, font_weight='bold', arrows=True, alpha=0.8, connectionstyle='arc3,rad=0.25', arrowsize=20)
plt.show()
```

```python
# Calculate in-degree and out-degree of node p
print(f"in-degree: {G_p.in_degree('p')}")
print(f"out-degree: {G_p.out_degree('p')}")
```

## 1.4.1.3 Communication
Figure 1.12: Strongly connected components of a digraph

```python
# Figure 1.12
G = nx.DiGraph()
G.add_edge(1, 1)
G.add_edge(2, 1)
G.add_edge(2, 3)
G.add_edge(3, 2)
```

```python
# Plots G
fig, ax = plt.subplots()
nx.draw_spring(G, ax=ax, node_size=500, with_labels=True, font_weight='bold', arrows=True, alpha=0.8, connectionstyle='arc3,rad=0.25', arrowsize=20)
plt.show()
```

```python
# Compute strongly connected components of G
print(f"Strongly connected components of G: {list(nx.strongly_connected_components(G))}")
```

## 1.4.1.4 Aperiodicity
Figure 1.13: A periodic digraph

```python
# Figure 1.13
edge_list = [
    ('a', 'b'),
    ('b', 'a'), ('b', 'c'),
    ('c', 'b'), ('c', 'd'),
    ('d', 'c'),
]

# Create DiGraph object
G = nx.DiGraph()
G.add_edges_from(edge_list)
```

```python
# Plots G
fig, ax = plt.subplots()
nx.draw(G, ax=ax, node_size=500, with_labels=True, font_weight='bold', arrows=True, alpha=0.8, connectionstyle='arc3,rad=0.25', arrowsize=20)
plt.show()
```

```python
# Check if graph G is aperiodic
print(f"Is G aperiodic?\n>>> {nx.is_aperiodic(G)}")
```

## 1.4.1.5 Adjacency Matrices

```python
# Define the adjacency matrix: A
A = ((1, 0, 0),
     (1, 1, 1),
     (1, 1, 1))

# Convert A to NumPy array
A = np.array(A)
# Create DiGraph object
G = qe.DiGraph(A)
```

```python
# Compute strongly connected components of G
print(f"Strongly connected components of G: {G.strongly_connected_components}")
```
